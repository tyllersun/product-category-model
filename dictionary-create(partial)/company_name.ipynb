```

def append_dataframe(new_data_name, data):
  new_data = pd.read_csv(new_data_name)
  #print(new_data.head())
  new_data = pd.DataFrame(new_data)
  data = data.append(new_data)
  #print(data.head())
  return data
```

```
def replace_word(word_list, text):
  for word in word_list:
    text = text.replace(str(word), "")
  return text
```
