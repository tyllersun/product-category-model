In the remove function, we remove number and punctation(both in Chinese and Englist) 


```

import re
 
def remove(text):
    remove_chars = '[0-9’!"#$%&\'()*+,-./:;<=>?@，。?★、…【】《》？“”‘’！[\\]^_`{|}~]+'
    return re.sub(remove_chars,'', text)
```

In this block, we use nltk to help us find out the noun the the text



```
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def find_noun(lines):
  word = ""
  # function to test if something is a noun
  is_noun = lambda pos: pos[:2] == 'NN'
  # do the nlp stuff
  tokenized = nltk.word_tokenize(lines)
  nouns = [word for (word, pos) in nltk.pos_tag(tokenized) if (is_noun(pos) and len(pos) > 1)]
  print(nouns)
  for i in range(len(nouns)):
    print(nouns[i])
    word = word + " " + nouns[i]
  return word
```

```
lines = "dog ET eqwe"
print(find_noun(lines))
```

In this part, I use nltk to drop out the non-English word

```
nltk.download('words')
def delete_non_English(sent):
  words = set(nltk.corpus.words.words())
  return " ".join(w for w in nltk.wordpunct_tokenize(sent) \
          if w.lower() in words or not w.isalpha())
# 'Io to the beach with my'
```

```
sent = "Io andiamo to the beach with my amico."  #test the function
print(delete_non_English(sent))
```

```
def delete_word_less_than_2(anytext):
  total = anytext.split(" ")
  need = []
  for i in range(len(total)):
    if len(total[i]) > 2:
      need.append(total[i])
  word = ""
  for i in range(len(need)):
    #print(need[i])
    word = word + " " + need[i]
  return word
```

```
for i in range(len(data)):
  data.loc[i,"noun"] = delete_word_less_than_2(data.loc[i,6])
  print(data.loc[i,"noun"])
```

```
def get_last_word(word):
  word = word.split(" ")[-1]
  return word
```

```
for i in range(len(data)):
  data.loc[i,"last word"] = get_last_word(data.loc[i,"noun"])
```
